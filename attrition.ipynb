## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers
from tensorflow.keras.layers import Input, Dense

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
```

```python
# Create a list of at least 10 column names to use as X data
selected_columns = ['Education', 'Age', 'DistanceFromHome', 'JobSatisfaction', 'OverTime', 'StockOptionLevel',
                    'WorkLifeBalance', 'YearsAtCompany', 'YearsSinceLastPromotion', 'NumCompaniesWorked']

# Create X_df using your selected columns
X_df = attrition_df[selected_columns]

# Show the data types for X_df
print(X_df.dtypes)
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.3, random_state=69)
```

```python
# Convert X data to numeric
X_train['OverTime'] = X_train['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0)
X_test['OverTime'] = X_test['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_train_scaled = scaler.fit_transform(X_train)

# Scale the training and testing data
X_test_scaled = scaler.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
encoder_dept = OneHotEncoder(sparse=False)

# Fit the encoder to the training data
encoder_dept.fit(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_dept_encoded = encoder_dept.transform(y_train[['Department']])
y_test_dept_encoded = encoder_dept.transform(y_test[['Department']])

print("Encoded Department - Training Data:")
print(y_train_dept_encoded[:5])
print("Encoded Department - Testing Data:")
print(y_test_dept_encoded[:5])
```

```python
# Create a OneHotEncoder for the Attrition column
encoder_attrition = OneHotEncoder(sparse=False)

# Fit the encoder to the training data
encoder_attrition.fit(y_train[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_attrition_encoded = encoder_attrition.transform(y_train[['Attrition']])
y_test_attrition_encoded = encoder_attrition.transform(y_test[['Attrition']])

print("Encoded Attrition - Training Data:")
print(y_train_attrition_encoded[:5])
print("Encoded Attrition - Testing Data:")
print(y_test_attrition_encoded[:5])
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
n_features = X_train_scaled.shape[1]

# Create the input layer
input_layer = Input(shape=(n_features,), name='input')

# Create at least two shared layers
shared1 = Dense(64, activation='relu', name='shared1')(input_layer)
shared2 = Dense(128, activation='relu', name='shared2')(shared1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
# Create the hidden layer
department_hidden = Dense(32, activation='relu', name='department_hidden')(shared2)

# Create the output layer
department_output = Dense(y_train_dept_encoded.shape[1], activation='softmax', name='department_output')(department_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer
# Create the hidden layer
attrition_hidden = Dense(32, activation='relu', name='attrition_hidden')(shared2)

# Create the output layer
attrition_output = Dense(2, activation='sigmoid', name='attrition_output')(attrition_hidden)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[department_output, attrition_output], name='attrition_model')

# Compile the model
model.compile(optimizer='adam',
              loss={'department_output': 'categorical_crossentropy', 'attrition_output': 'binary_crossentropy'},
              metrics={'department_output': 'accuracy', 'attrition_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
history = model.fit(
    x=X_train_scaled,
    y={'department_output': y_train_dept_encoded, 'attrition_output': y_train_attrition_encoded},
    batch_size=32,
    epochs=100,
    verbose=1,
    validation_data=(X_test_scaled, {'department_output': y_test_dept_encoded, 'attrition_output': y_test_attrition_encoded})
)
```

```python
# Attempt to train the model again
history = model.fit(
    x=X_train_scaled,
    y={'department_output': y_train_dept_encoded, 'attrition_output': y_train_attrition_encoded},
    batch_size=32,
    epochs=100,
    verbose=1,
    validation_data=(X_test_scaled, {'department_output': y_test_dept_encoded, 'attrition_output': y_test_attrition_encoded})
)
```

```python
# Evaluate the model with the testing data
# Evaluate the model on the testing data
evaluation_results = model.evaluate(
    x=X_test_scaled,
    y={'department_output': y_test_dept_encoded, 'attrition_output': y_test_attrition_encoded},
    verbose=1
)
```

```python
# Print the accuracy for both department and attrition
department_accuracy = evaluation_results[-2]
attrition_accuracy = evaluation_results[-1]

print(f"Department predictions accuracy: {department_accuracy}")
print(f"Attrition predictions accuracy: {attrition_accuracy}")
```

1. Is accuracy the best metric to use on this data? Why or why not?

Accuracy can be a useful metric for getting a general sense of model performance, especially when the classes in the dataset are balanced. However, it might not be the best metric in all cases, particularly for imbalanced datasets where one class significantly outnumbers the other.

For the 'Department' classification, if the distribution of classes is relatively balanced, accuracy might give a reasonable measure of performance. However, if some departments are much less common than others, accuracy could be misleading.

For the 'Attrition' prediction, if the number of employees leaving is much lower than those staying, then accuracy might not reflect the model's ability to correctly identify the less frequent class. In such cases, other metrics like precision, recall, F1 score, and ROC AUC could provide more insight into model performance, especially for identifying true positives and handling false negatives.

2. What activation functions did you choose for your output layers, and why?

For the 'Department' output, a softmax activation function was chosen because it's a multi-class classification problem. Softmax is suitable for these types of problems because it converts the output scores from the model into probabilities by comparing the score for each class relative to the scores for other classes. This allows the model to output a probability distribution over the classes.

For the 'Attrition' output, initially, a sigmoid activation function was suggested, typical for binary classification tasks, as it outputs a probability score between 0 and 1, indicating the likelihood of the sample belonging to a positive class. To accommodate the one-hot encoded 'Attrition' data, softmax was used. This was suitable because it also handles the 'Attrition' prediction as a case of multi-class classification with two classes.
